```python
import pandas as pd

df_subtitle=pd.read_csv('subtitle_data.csv')

df_subtitle

def custom_chunker(chunk_size, unique_id, content, name, chunk_buffer):
    # Split content into tokens
    tokens = [token for token in content.split()]
    # Get the length of the tokens
    num_tokens = len(tokens)

    # Initialize index and count
    start_index, chunk_count = 0, 0

    while True:
        if start_index == 0:
            chunk_start, chunk_end = start_index,chunk_buffer
            if chunk_end > num_tokens:
                new_id.append(f"{unique_id}-{chunk_count}")
                new_content.append(" ".join(tokens[chunk_start:]))
                names.append(name)  # Append the name here
                break
            else:
                new_id.append(f"{unique_id}-{chunk_count}")
                new_content.append(" ".join(tokens[chunk_start:chunk_end]))
                names.append(name)  # Append the name here
                start_index = chunk_end
        else:
            chunk_start, chunk_end = start_index - chunk_size, start_index + chunk_buffer - chunk_size
            if chunk_end > num_tokens:
                new_id.append(f"{unique_id}-{chunk_count}")
                new_content.append(" ".join(tokens[chunk_start:]))
                names.append(name)  # Append the name here
                break
            else:
                new_id.append(f"{unique_id}-{chunk_count}")
                new_content.append(" ".join(tokens[chunk_start:chunk_end]))
                names.append(name)  # Append the name here
                start_index = chunk_end
        chunk_count += 1
```

```python
new_id = []
new_content = []
names = []

for i in range(len(df_subtitle)):
    custom_chunker(25, i, df_subtitle['subtitle_processed'][i], df_subtitle['name'][i], 250)
```

```python
new_df = pd.DataFrame({'id': new_id, 'subtitle_content': new_content, 'name': names})
new_df
```
